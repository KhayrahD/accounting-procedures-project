```python
# total sales function

def total_sales():
    """
    Calculate total sales for the day based on morning and evening sales data.
    
    """
    try:
        # user to enter morning sales data and convert it to a list of floats
        morning_sales = list(map(float, input("Enter morning sales data (comma-separated): ").split(',')))

        # user to enter evening sales data and convert it to a list of floats
        evening_sales = list(map(float, input("Enter evening sales data (comma-separated): ").split(',')))

        # to calculate the total sales for the day by summing morning and evening sales, rounded to 2 decimal places
        daily_sales = round(sum(morning_sales) + sum(evening_sales), 2)

        # to return the calculated total sales for the day
        return daily_sales

    except ValueError:
        # to handle a case where the user enters invalid input (non-numeric values)
        print("Invalid input. Please enter numbers separated by a comma.")

        # to return 0 if there's an error
        return 0
    
    
    
    
# worker salary function

def worker_salary():
    """
    Calculate Worker's Salary based on given hourly rate and hours worked by a worker

    """
    try:
        # user to enter the hourly rate for workers and convert it to a float
        hourly_rate = float(input("Enter hourly rate for workers in Naira: "))

        # user to enter the number of hours the worker worked and convert it to a float
        hours_worked = float(input("Enter hours the worker worked for: "))

        # to calculate the worker's salary by multiplying the hourly rate with the hours worked, rounded to 2 decimal places
        salary = round(hourly_rate * hours_worked, 2)

        # to return the calculated worker's salary
        return salary

    except ValueError:
        # to handle case where the user enters invalid input like non-numeric values
        print("Invalid input. Please enter numbers for hourly rate and hours worked.")

        # to return 0 if there's an error
        return 0
    
    
    
    
# profit/loss calculation function

def profit_loss_cal():
    """
    Calculate Profit: given a list of numbers representing total sales and total cost of items sold.
    
    """
    try:
        # user to enter the total sales and convert it to a float
        total_sales = float(input("Enter total sales: "))

        # user to enter the total cost and convert it to a float
        total_cost = float(input("Enter total cost: "))

        # to calculate the difference between total sales and total cost, rounded to 2 decimal places
        amount_difference = round(total_sales - total_cost, 2)

        # to determine if it's a profit or loss
        if amount_difference > 0:
            # If there's a profit, set profit to the calculated amount and loss to 0
            profit = amount_difference
            loss = 0
        elif amount_difference < 0:
            # If there's a loss, set loss to the absolute value of the calculated amount and profit to 0
            profit = 0
            loss = abs(amount_difference)
        else:
            # If there's neither profit nor loss, set both profit and loss to 0
            profit = 0
            loss = 0

        # to return the calculated profit and loss
        return profit, loss

    except ValueError:
        # to handle the case where the user enters invalid input (non-numeric values)
        print("Invalid input. Please enter numbers for total sales and total cost.")

        # to return 0 for both profit and loss if there's an error
        return 0, 0
    
    
    
    
# shift tips calculation function

def shift_tips():
    """
    Calculate Tips for a Shift: from sales data for a specific shift, workers get tipped for the shift (2% of shift sales).

    """
    try:
        # user to enter sales data for the shift and convert it to a list of floats
        shift_sales = list(map(float, input("Enter sales data for the shift (comma-separated): ").split(',')))

        # user to enter the tip percentage for workers and convert it to a float
        tip_percentage = float(input("Enter tip percentage for workers: "))

        # to check if the tip percentage is zero, and handle it as an invalid input
        if tip_percentage == 0:
            print("Invalid input. Tip percentage cannot be zero.")
            
            # to return 0 if there's an error
            return 0
        else:
            # to calculate the shift tip by multiplying the sum of shift sales with the tip percentage
            #rounded to 2 decimal places
            shift_tip = round(sum(shift_sales) * tip_percentage, 2)
            
            # to return the calculated shift tip
            return shift_tip

    except ValueError:
        # to handle the case where the user enters invalid input (non-numeric values for sales data)
        print("Invalid input. Please enter numbers for sales data.")
        
        # to return 0 if there's an error
        return 0
    
    
    
    
# total tips calculation function

def total_tips():
    """
    Calculate Total Tips for the Day: with sales data for morning and evening shifts

    """
    try:
        # user to enter morning sales data and convert it to a list of floats
        morning_sales = list(map(float, input("Enter morning sales data (comma-separated): ").split(',')))

        # user to enter evening sales data and convert it to a list of floats
        evening_sales = list(map(float, input("Enter evening sales data (comma-separated): ").split(',')))

        # user to enter the tip percentage for workers and convert it to a float
        tip_percentage = float(input("Enter tip percentage for workers: "))

        # Check if the tip percentage is zero, and handle this as an invalid input
        if tip_percentage == 0:
            print("Invalid input. Tip percentage cannot be zero.")
            
            # to return 0 if there's an error
            return 0
        else:
            # to calculate the tips for the morning and evening shifts
            morning_tip = sum(morning_sales) * tip_percentage
            evening_tip = sum(evening_sales) * tip_percentage
            
            # to calculate the total tips for the day by summing the tips from both shifts, rounded to 2 decimal places
            total_tip = round(morning_tip + evening_tip, 2)
            
            # to return the calculated total tips for the day
            return total_tip

    except ValueError:
        # to handle the case where the user enters invalid input (non-numeric values for sales data)
        print("Invalid input. Please enter numbers for sales data.")
        
        # to return 0 if there's an error
        return 0
    
    
    
    
# display menu function

def display_menu():
    """
    A user-friendly interface that displays a menu of available operations.
    It prompts the user to enter desired numbers within 1 to 6 for a specific operation.
    It displays the result of the selected operation clearly to the user.
    It also allows the user to perform multiple calculations until choosing to exit the system.
    
    """
    while True:
        # to display the welcome message and menu options
        print("\nWelcome to Tolu's Cuisine Automated Accounting System")
        print('1. Calculate Total Sales')
        print('2. Calculate Worker Salary')
        print('3. Calculate Profit or Loss')
        print('4. Calculate Shift Tips')
        print('5. Calculate Total Tips')
        print('6. Exit')

        # to prompt the user to enter their choice (1-6)
        choice = input('Enter your choice (1-6): ')

        # to check the user's choice and perform the corresponding operation
        if choice == '1':
            # to call the total_sales function and display the result
            sales_total = total_sales()
            print(f"Total Sales for the Day is {sales_total} naira")
        elif choice == '2':
            # to call the worker_salary function and display the result
            salary_for_worker = worker_salary()
            print(f"Total Worker's Salary is {salary_for_worker} naira")
        elif choice == '3':
            # to call the profit_loss_cal function and display the result
            profit_acquired, loss_incurred = profit_loss_cal()
            if profit_acquired == 0:
                print(f"Total Loss incurred is {loss_incurred} naira")
            elif loss_incurred == 0:
                print(f"Total profit made is {profit_acquired} naira")
            else:
                print("Neither profit was made nor loss was incurred.")
        elif choice == '4':
            # to call the shift_tips function and display the result
            tips_for_shift = shift_tips()
            if tips_for_shift == 0:
                None  # to do nothing if there's an error in shift_tips
            else:
                print(f"Total tips for this shift is {tips_for_shift} naira")
        elif choice == '5':
            # to call the total_tips function and display the result
            tips_total = total_tips()
            if tips_total == 0:
                None  # to do nothing if there's an error in total_tips
            else:
                print(f"Total Tips for the Day is {tips_total} naira")
        elif choice == '6':
            # to display exit message and break out of the loop to end the program
            print("You can now exit the system, thank you.")
            break
        else:
            # to  display an error message for invalid choices
            print('Invalid choice. Please enter a number between 1 and 6.')

# to run the program by calling the display_menu function
display_menu()
```
